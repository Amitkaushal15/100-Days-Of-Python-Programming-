## While Loop

#### A Python while loop repeatedly executes a block of code as long as a specified condition remains True. It is primarily used when the exact number of iterations is unknown before the loop begins.

```python
# while
# for
```

```python
while condition:
    code
```

```python
number = int(input("Enter the number: "))
i = 1
while i < 11:
    print(number, "*", i, "=", number * i)
    i += 1
             
```
