## Operators

#### Operators are used to perform operations on variables and values. Python has the following operators:
#### 1. Arithmetic operators 
#### 2. Comparison operators
#### 3. Logical operators
#### 4. Bitwise operators
#### 5. Assignment operators
#### 6. Identity operators
#### 7. Membership operators

### 1. Arithmetic operators

```python
a = 6
b = 9
print(a+b)
```

```python
print(a-b)
```

```python
print(a*b)
```

```python
print(a/b)
```

```python
print(a%b)
```

```python
print(a**b)
```

```python
print(b//a)
```

```python
print(a>b)
```

```python
print(a<b)
```

```python
print(a>=b)
```

```python
print(a<=b)
```

### 3. Logical operators

```python
x = True
y = False

print(x or y)
```

```python
print(x and y)
```

```python
print(not x)
```

```python
print(not y)
```

### 4. Bitwise operators

```python
x = 10
y = 8
print(x & y)
```

```python
print(x | y)
```

```python
print(y >> 2)
```

```python
print(x << 8)
```

```python
print(~x)
```

### 5. Assignment operators

```python
a = 5
print(a)
```

```python
a += 4
# a = a + 4
print(a)
```

```python
a -= 3
print (a)
```

```python
a *= 6
print (a)
```

```python
a /= 5
print (a)
```

### 6. Identity operators

```python
a = 4
b = 4
print(a is b)
```

```python
a = "Hello"
b = "Hello"
print(a is b)
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)
```

```python
a = "Hello-World"
b = "Hello-World"
print(a is not b)
```

### 7. Membership operators

```python
x = "Patna"
print("P" in x)
```

```python
x = "Patna"
print("P" not in x)
```

```python
x = [ 1, 2, 3]
print(1 in x)
```

```python
x = [ 1, 2, 3]
print(5 in x)
```
