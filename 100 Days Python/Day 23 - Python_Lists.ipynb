## Code Examples

### 1. Data Type Constraints

#### Lists allow you to mix structural data types smoothly, while arrays reject mixed-type mutations or force upcasting.

```python
# A list can hold anything
my_list = [10, "Hello", 3.14]
print(my_list)

# An array must be uniform
import array as arr
my_array = arr.array('i', [1, 2, 3, 4])  # 'i' specifies signed integers
print(my_array)
```

### 2. Mathematical Performance

#### If you multiply a list, Python repeats the sequence. If you multiply an array, it scales the numerical contents inside

```python
# List multiplication replicates elements
print([1, 2, 3] * 2) 
# Output: [1, 2, 3, 1, 2, 3]
```

```python
# NumPy Array multiplication performs vector math
import numpy as np
np_arr = np.array([1, 2, 3])
print(np_arr * 2) 
# Output: [2, 4, 6]
```
