## Identifiers
#### - A python identifier is a name used to identity a variable, function, class, module or other object.

### Rules for setting Identifiers
#### - Can only start with an alphabet or _
#### - Followed by 0 or more letter, _ and digits
#### - keywords cannot be used as an identifiers


```python
name = "Amit"
print(name)
```

```python
_ = "Kaushal"
print(_)
```
