# Python Data Types Categories

## Basic Types:

### * Int: Whole nums (1,2,3)
### * Float: Decimals (3.14, 9.5)
### * Complex: Real+Imag (1+2j)
### * Bool: True/False
### * Str: Text ("Hello")

# Container Types:

#### * List: Ordered items
####    [1, 2, 3]
#### * Tuple: Ordered, immutable
####    (1, 2, 3)
#### * Set: Unordered, Unique
####   {1, 2, 3}
#### * Dist: Key-Value Pairs
####   {'Key1' : 'value1', 'Key2' : 'value2'}

# User-Defined Types:
#### * Class: Obj blueprint

```python
# Large Numbers
print(1e308) # 1 * 10^308
print(1e309) # 'inf'(exceeds float limit)
```

### Float

```python
# Floating-point numbers
print(5.9) # Standard

# Large Floating-point numbers
print(1.8e307) # Large
print(1.8e308) # Larger
print(1.8e309) # Too Large (inf)
```

### bool

```python
# Boolean Values
print(True)
print(False)
```

### complex

```python
# complex numbers
print(2+9j)
```

### str

```python
# String with different quotation styles
print('Amit') # single quotes
print("Kaushal") # Double Quotes
print('''Navneet ''') # triple single quotes
print("""Adrija """)  #triple double quotes
```

### list

```python
a = ["Amit", 9 , 8.9, True]
print(a)
print([6, 2.9, "Kaushal", True])
```

### tuple

```python
# Tuple
print((9, 8, 7, 6, 5))
a = (1, 2, 3, 4, 5, 6)
print(a)
```

### set

```python
# Set 
print({ 1, 2, 3, 4, 5})
print({ "Amit", 1, 3.5, True})
a = { 1, 2, 3, 4, 5}
print(a)
```

### dict

```python
# Dictionary
a = { '1' : 'Amit', '2' : 'Kaushal', 'age' : '25'}
print (a)
```
