## Break Continue and pass statements

#### The break, continue, and pass statements are fundamental control flow tools used in loops and conditionals. Break exits the loop entirely, continue skips the current iteration and jumps to the next, and pass acts as a temporary, do-nothing placeholder.

### 1. break: Exit a Loop

#### Example:

```python
for i in range(1, 10):
    if i == 5:
        break # Stops the loop entirely when i hits 5
    print(i)
```

### 2. continue: Skip an Iteration

#### Example:

```python
for i in range(1, 6):
    if i == 3:
        continue # Skips printing '3' and jumps straight to '4'
    print(i)
```

### 3. pass: Do Nothing (Placeholder)

#### Example:

```python
for i in range(1, 4):
    if i == 2:
        pass # Placeholder for future logic, loop continues normally
    print(i)
```
