## For Loop

#### A for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each item in that sequence. Unlike other languages that use counter variables, Python loops directly over the items of an iterable object.

### Basic Syntax

```python
for item in sequence:
    # Code to repeat for each item
```

#### for: Keyword that starts the loop.
#### item: A temporary variable that holds the current element during this round of the loop.
#### in: Keyword linking the variable to your sequence.
#### sequence: The collection you want to look through (e.g., a list of names or numbers).
#### Indentation: The code block inside the loop must be indented (usually 4 spaces)

### Common Examples

### 1. Looping Through a List

#### You can step through every item in a list from start to finish.

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

### 2. Repeating Code a Fixed Number of Times

#### Use the built-in range() function to loop a specific number of times.

```python
# range(5) generates numbers from 0 up to (but not including) 5
for i in range(5):
    print("Hello!")
```

#### If you do not need to use the number variable inside your loop, it is common Python practice to use an underscore (_) as a throwaway variable name:

```python
for _ in range(3):
    print("Repeat this text")
```

### 3. Looping Through a String

#### Strings are sequences of individual characters, so the loop steps through them letter by letter.

```python
for letter in "Python":
    print(letter)
```
