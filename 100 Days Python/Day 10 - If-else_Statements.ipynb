## Decision Control

#### Scenario - 1

```python
# correct email - offdefsecurity@gmail.com
# correct password - 1234

email = input("Enter the your email: ")
password = input("Enter the Password: ")
if email == "offdefsecurity@gmail.com" and password == "1234":
    print("Welcome")
else:
    print("Incorrect credentials")
```

#### Scenario - 2

```python
# correct email - offdefsecurity@gmail.com
# correct password - 1234

email = input("Enter the your email: ")
password = input("Enter the Password: ")
if email == "offdefsecurity@gmail.com" and password == "1234":
    print("Welcome")
elif email == "offdefsecurity@gmail.com" and password != "1234":
    print("Incorrect password")
    password = input("Enter password again")
    if password == "1234":
        print("Finally correct")
    else:
        print("Still Incorrect")
else:
    print("Incorrect credentials")
```

#### Scenario - 3

```python
email = input("Enter the your email: ")
if '@' in email:
    password = input("Enter the Password: ")
    if email == "offdefsecurity@gmail.com" and password == "1234":
        print("Welcome")
    elif email == "offdefsecurity@gmail.com" and password != "1234":
        print("Incorrect password")
        password = input("Enter password again")
        if password == "1234":
            print("Finally correct")
        else:
            print("Still Incorrect")
    else:
        print("Incorrect credentials")
else:
    print("Incorrect Email")
```
