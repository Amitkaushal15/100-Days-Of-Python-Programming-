#### 6. pow

#### The Built-in pow() Function

```python
# Returns 8 (integer)
result = pow(2, 3)  
print(result)
```

```python
# Calculates (5 ** 3) % 100 -> 125 % 100 -> Returns 25
result_mod = pow(5, 3, 100)  
print(result_mod)
```

```python
import math

# Returns 8.0 (float)
result = math.pow(2, 3) 
print(result)
```

#### 9. divmod

#### Syntax

```python
divmod(dividend, divisor)
```

```python
# Returns a tuple of (quotient, remainder)
result = divmod(10, 3)
print(result)  
# Output: (3, 1)

# You can cleanly unpack the tuple directly into variables
quotient, remainder = divmod(10, 3)
print(quotient)   # Output: 3
print(remainder)  # Output: 1
```

#### 10. bin/oct/hex

#### Overview of Functions

#### Code Example

```python
number = 255

# Basic conversions
binary_str = bin(number)  # '0b11111111'
octal_str = oct(number)   # '0o377'
hex_str = hex(number)   # '0xff'

print(binary_str)
print(octal_str)
print(hex_str)
```

#### 14. sum

```python
numbers = [1, 2, 3, 4, 5]
total = sum(numbers)
print(total)  # Output: 15
```

```python
# Summing a tuple
tuple_sum = sum((10.5, 20.5))  # Output: 31.0

# Summing a range of numbers (e.g., 1 to 100)
range_sum = sum(range(1, 101))  # Output: 5050

# Summing dictionary VALUES (must specify .values())
student_marks = {"Math": 90, "Science": 85}
marks_sum = sum(student_marks.values())  # Output: 175
```
