#### 5. format

```python
"Hello, My name is {} and  I am {} years old".format("Amit", 24)
```

```python
"Hello, My name is {1} and  I am {0} years old".format("Amit", 24)
```

```python
"Hello, My name is {0} and  I am {1} years old".format("Amit", 24)
```

```python
"Hello, My name is {name} and  I am {age} years old".format(name = "Amit", age = 24)
```
