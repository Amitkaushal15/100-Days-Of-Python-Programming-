## Operation On Strings

- Arithmetic Operations (+, *)
- Relational Operations
- Logical Operations
- Loops on Strings
- Membership Operations

### Arithmetic Operations (+, * only)

```python
"Hello" + "-" + "World!"   #concatenation
```

```python
print("*" * 50)
```

```python
print("#" * 40)
```

```python
print("Hello"" "*5)
```

### Relational Operations

```python
"Hello" == "World!"
```

```python
"Hello" != "World!"
```

```python
"Bihar" > "Pune"
# Lexiographically 
```

```python
"Bihar" < "Pune"
```

```python
"Goa" < "Kolkata"
```

```python
"goa" < "Goa"
```

### Logical Operations

```python
"hello" and "world"
```

```python
"" and "Hello"
```

```python
"" or "World"
```

```python
"Hello" or "World"
```

```python
print(not "Hello")
```

```python
not ""
```

### Loops on Strings

```python
c = "Hello World!"
for i in c:
    print(i)
```

```python
c = "Hello World!"
for i in c[0:5:1]:
    print(i)
```
