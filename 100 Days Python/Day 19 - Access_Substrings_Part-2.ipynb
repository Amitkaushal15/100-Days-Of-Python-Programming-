## Accessing Substrings from a string

```python
# concept of Indexing
c = "Hello World!"
print(c)
```

```python
print(c[6])
```

```python
print(c[12])
```

```python
print(c[11])
```

```python
# Types of Indexing
# Positive Indexing
# Negative Indexing
c = "Hello Amit"
print(c[6])
print(c[-1])
```

```python
print(c[0])
```

```python
print(c[-4])
```

```python
# Slicing
c = "Amit Kaushal"
print(c)
```

```python
print(c[0:4])
```

```python
print(c[4:])
```

```python
print(c[3:])
```

```python
print(c[:5])
```

```python
print(c[2:8])
```

```python
print(c[:])
```

```python
print(c[2:8:2])
```

```python
print(c[0:8:3])
```

```python
print(c[0:6:-1]) # we can't take steps in negative number
```

```python
print(c[-5:-1:2])
```

```python
print(c[-8:-0:3])
```

```python
print(c[-8:-1:3])
```

```python
print(c[::-1])  # use for reverse
```

```python
print(c[-1:-8:-1])
```
