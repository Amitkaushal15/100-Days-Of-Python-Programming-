## Dictionary
#### A Python dictionary is a built-in data structure used to store data as key: value pairs. Unlike lists that look up items by numerical indexes, dictionaries look up items using unique, descriptive keys. As of Python 3.7, dictionaries are ordered collections, meaning they maintain the exact order in which you insert items.

- Rule 1: Dictionary has no indexing
- Rule 2: Dictionary is a mutable types
- Rule 3: Keys -> immutable, value -> they can be mutable
- Rule 4: Keys should be unique

* Mutable -> List/Sets/Dictionary
* Immutable -> Strings/Tuples/int/float/bool/complex
