### Statics vs Dynamic Typing:
#### * Dynamic Typing(Python, PHP): Types change during execution.
#### * Static Typing (C, C++, Java): Fixed types before runtime

### Statics vs Dynamic Binding:
####   * Static Binding: Types fixed at compile time.
####   * Dynamic Binding: Types associated at runtime, changeable during execution.
