## Type Conversion
### Type conversion in Python is the process of changing a value or variable from one data type to another.

## Python categorizes type conversion into two primary types:
#### 1. Implicit Type Conversion
#### 2. Explicit Type Conversion

### 1. Implicit Type Conversion (Automatic)

#### Implicit type conversion happens automatically when the Python interpreter executes code and encounters mixed data types.
#### Behavior: Python automatically promotes a "lower" data type (like an integer) to a "higher" data type (like a float).
#### Purpose: This protects your program from accidental data loss or precision errors.
#### Example: Adding an integer to a float results in a float.

```python
num_int = 10       # int type
num_float = 5.5    # float type

# Python automatically converts num_int to a float before adding
result = num_int + num_float

print(result)       # Output: 15.5
print(type(result)) # Output: <class 'float'>
```

### 2. Explicit Type Conversion (Type Casting)

#### Explicit type conversion happens manually when the programmer explicitly forces a value into a specific data type.
#### Behavior: You wrap your variable inside one of Python's built-in conversion functions.
#### Purpose: It is used when Python cannot automatically combine incompatible types (like a string and an integer), which would normally throw a TypeError.
#### Risk: Enforcing a conversion can sometimes lead to data loss (e.g., converting 5.9 into an integer truncates the decimal entirely to become 5)

```python
# Converting a string to an integer
age_str = "25"
age_int = int(age_str) 

print(age_int + 5)  # Output: 30
```
