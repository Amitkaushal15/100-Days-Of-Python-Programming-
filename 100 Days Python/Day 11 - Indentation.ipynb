## Indentation 

#### Indentation in Python refers to the whitespaces (spaces or tabs) at the beginning of a line of code. Unlike languages like C, Java, or JavaScript that use curly braces {} to group code, Python uses mandatory indentation to define blocks of code and scope.

### 📌 How It Works

#### When you write control structures like functions, loops, or conditionals, you must end the statement with a colon (:). Every subsequent line belonging to that structure must be shifted to the right by the same number of spaces.

```python
# Correct Example
if 5 > 2:
    print("This line is indented!")  # Runs only if the condition is true
print("This line is not indented.")  # Always runs (outside the block)
```

### 📏 Essential Rules

#### - Standard practice: The official Python style guide, PEP 8, recommends using 4 spaces per indentation level.
#### - Consistency: You can technically use any number of spaces, but every line within the same block must use the exact same alignment.
#### - Never mix tabs and spaces: Mixing them will confuse the interpreter and throw a TabError or IndentationError.
First line rule: You cannot add indentation to the very first line of a Python script.
